#Condicional / conversão para inteiro


```python
idade = 15

def verifica(idade):
  if idade >= 18:
    print('Tem permissão para dirigir')
  else:
    print('Não tem permissão para dirigir')

verifica(idade)
```

```python
def verifica_sem_parametros():
  idade = input('Qual sua idade? ')
  idade = int(idade)
  if idade >= 18:
    print('Tem permissão para dirigir')
  else:
    print('Não tem permissão para dirigir')

verifica_sem_parametros()
```

#Laços e loops

```python
idades = [18, 22, 15, 50]
```

```python
#for fora da função

def verifica(idade):
  if idade >= 18:
    print(f'{idade} anos de idade, TEM permissão para dirigir')
  else: 
    print(f'{idade} anos de idade, NÃO TEM permissão para dirigir')

for idade in idades:
  verifica(idade)
```

```python
#for dentro da função

def verifica(idades):
  for idade in idades:

    if idade >= 18:
      print(f'{idade} anos de idade, TEM permissão para dirigir')
    else: 
      print(f'{idade} anos de idade, NÃO TEM permissão para dirigir')

verifica(idades)
```

```python
permissoes = []
idades = [20, 14, 40]

def verifica (idades, permissoes):
  for idade in idades:
    if idade >= 18:
      permissoes.append(True)
    else:
      permissoes.append(False)

verifica (idades, permissoes)
permissoes
```

```python
for permissao in permissoes:
  if permissao == True:
    print('Tem permissão para dirigir')
  else:
    print('Não tem permissão para dirirgir')
```
